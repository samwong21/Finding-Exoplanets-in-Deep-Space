```python
from sklearn import preprocessing
from sklearn.pipeline import make_pipeline
from sklearn import tree
```

```python
X = exoTrain.drop(['LABEL'], axis=1) # drop 'LABEL' column
```

```python
y = exoTrain['LABEL']
```

```python
def fit_tree(X, y, d):
    T = tree.DecisionTreeClassifier(max_depth=d)
    T.fit(X, y)
    return T
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
T = fit_tree(X_train, y_train, d=2)
print(T.score(X_train, y_train), T.score(X_test, y_test))

fig, ax = plt.subplots(1, figsize = (10, 10))
p = tree.plot_tree(T, filled=True, feature_names =X.columns)
plt.show()
```

```python
fig, ax = plt.subplots(1, figsize= (12, 8))
depths = range(1, 10)
train_scores = []
test_scores = []

for d in depths:
    T = fit_tree(X_train, y_train, d)
    train_score = T.score(X_train, y_train)
    test_score = T.score(X_test, y_test)
    
    train_scores.append(train_score)
    test_scores.append(test_score)

ax.scatter(depths, train_scores, label='train score')
ax.scatter(depths, test_scores, label='test score')
ax.set(xlabel="depth", ylabel= 'score')
plt.legend(fontsize=20)
plt.show()
```

```python
fig, ax = plt.subplots(1, figsize= (12, 8))
depths = range(20, 30)
train_scores = []
test_scores = []

for d in depths:
    T = fit_tree(X_train, y_train, d)
    train_score = T.score(X_train, y_train)
    test_score = T.score(X_test, y_test)
    
    train_scores.append(train_score)
    test_scores.append(test_score)

ax.scatter(depths, train_scores, label='train score')
ax.scatter(depths, test_scores, label='test score')
ax.set(xlabel="depth", ylabel= 'score')
plt.legend(fontsize=20)
plt.show()
```
